```python
#thêm 1 sản phẩm
def themsp():
    id = products[-1]["id"] + 1 if products else 1
    ten = input("Nhập tên sản phẩm: ")
    gia = int(input("Nhập giá sản phẩm; "))
    mota = input("Nhập mô tả sản phẩm: ")

    new_sp = {
        "id": id,
        "name": ten,
        "price": gia,
        "description": mota
    }
    products.append(new_sp)

#thêm nhiều sản phẩm
sanphammoi = [(4, "samsung", 2500, "samsung best"),
                (5, "Xiaomi 13", 1500, "xiaomi best")]

for id, ten, gia, mota in sanphammoi:
    products.append({
        "id": id,
        "name": ten,
        "price": gia,
        "description": mota
    })
products
```

```python
#tìm kiếm sản phẩm
def timkiem():
    ten = input("Nhập tên sp muốn tìm: ")
    found = False

    for p in products:
        if ten.lower() in p['name'].lower():
            print(f"Đã tìm thấy sản phẩm: {p['id']} - {p['name']}")
            found = True
        
    if not found:
        print(f"Không tìm thấy {p['id']}")
timkiem()
```

```python
def capnhat(id):
    found = False
    for p in products:
        if p['id'] == id:
            name = input("Nhập tên mới")
            price = input("Nhập giá mới")
            description = input("Nhập mô tả mới")

            if name:
                p['name'] = name
            if price:
                p['price'] = price
            if description:
                p['description'] = description
            print(f"Cập nhật thành công {id}")
            found = True
    if not found:
      print(f"Không tìm thấy mã {id}")
capnhat(4)
```
