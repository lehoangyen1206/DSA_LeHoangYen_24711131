**1. Kiểu dữ liệu cơ bản:**
- int, float: Các phép toán số học
- str: Xử lý chuỗi, các phương thức string
- bool: Logic và điều kiện

**2. Collection types:**
- **List**: Có thứ tự, có thể thay đổi, cho phép trùng lặp → Dùng khi cần thay đổi dữ liệu
- **Tuple**: Có thứ tự, KHÔNG thể thay đổi, cho phép trùng lặp → Dùng cho dữ liệu cố định

- **Dictionary**: Lưu trữ cặp key-value, có thể thay đổi → Dùng khi cần mapping

**3. Khi nào dùng gì:**
- Cần lưu danh sách có thể thay đổi → **List**
- Cần lưu dữ liệu không đổi (như tọa độ) → **Tuple**
- Cần loại bỏ phần tử trùng → **Set**
- Cần mapping (ví dụ: tên → điểm) → **Dictionary**

**Bài 1.10:** Tính tiền điện theo công thức bậc thang:
- 0-100 kWh: 2000 VNĐ/kWh
- 101-200 kWh: 2500 VNĐ/kWh
- 201-300 kWh: 3000 VNĐ/kWh
- Trên 300 kWh: 3500 VNĐ/kWh

````python
def tinhtiendien():
    kwh = float(input("Nhập số kwh tiêu thụ: "))
    S = 0
    if kwh <=100:
        S = kwh * 2000
    elif kwh <=200:
        S = (100 * 2000) + (kwh - 100) * 2500
    elif kwh <=300:
        S = (100 * 2000) + (100 * 2500) + (kwh - 200) * 3000
    else:
        S = (100 * 2000) + (100 * 2500) + (100 * 3000) + (kwh - 300) * 3500
    return S
print(f"Tổng tiền điện phải trả là: {tinhtiendien():,} VNĐ")
````

**Bài 2.1:** Cho list: `numbers = [5, 2, 9, 1, 7, 3]`

````python
numbers = [5, 2, 9, 1, 7, 3]

#tìm số lớn nhất
max_val = numbers[0]
for n in numbers:
    if n > max_val:
        max_val = n
print(f"Số lớn nhất là: {max_val}")

#tìm số nhỏ nhất
min_val = numbers[0]
for n in numbers:
    if n < min_val:
        min_val = n
print(f"Số nhỏ nhất là: {min_val}")

#tính tổng các số
tong = 0
for n in numbers:
    tong += n
print(f"Tổng các số là: {tong}")
````

**Bài 2.12:** Tìm dãy con liên tiếp có tổng lớn nhất (Kadane's Algorithm)

Ví dụ: `[-2, 1, -3, 4, -1, 2, 1, -5, 4]` → Kết quả: 6 (từ [4, -1, 2, 1])

````python
numbers = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
def kadane(arr):
    max_current = max_global = arr[0]
    for n in arr[1:]:
        max_current = max(n, max_current + n)
        if max_current > max_global:
            max_global = max_current
    return max_global

ket_qua = kadane(numbers)
print(f"Tổng lớn nhất của dãy con liên tiếp là: {ket_qua}")
````

**Bài 3.9:** Mã hóa Caesar Cipher (dịch mỗi ký tự đi k vị trí)

Ví dụ: "abc", k=2 → "cde"

````python
text = "abc"
k = 2
ketqua = ""

for char in text:
    if char.isalpha():
        start =  ord('a') if char.islower() else ord('A')
        newchar = chr((ord(char) - start + k) % 26 + start)
        ketqua += newchar
    else:
        ketqua += char
print(f"Kết quả: {ketqua}")
````

**Bài 3.10:** Tìm chuỗi con chung dài nhất (Longest Common Substring)

Ví dụ: "ABCDGH" và "ACDGHR" → "CDGH"

````python
def longest_common_substring(s1, s2):
    m, n = len(s1), len(s2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    max_len = 0
    end_pos = 0
    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = dp[i-1][j-1] + 1
                if dp[i][j] > max_len:
                    max_len = dp[i][j]
                    end_pos = i
            else:
                dp[i][j] = 0
                
    return s1[end_pos - max_len : end_pos]

s1, s2 = "ABCDGH", "ACDGHR"
print(f"Kết quả: {longest_common_substring(s1, s2)}")
````

Ví dụ: `[1, 2, 2, 3, 4, 4, 5]` → `{1, 2, 3, 4, 5}`

**Bài 5.2:** Thêm và xóa phần tử trong set

````python
set1 = {1,2,2,3,4,4,5}
#thêm
set1.add(7)
print(f"Thêm: {set1}")

#xóa
set1.remove(3)
print(f"Xóa: {set1}")
````

### CẤP ĐỘ 2: TRUNG BÌNH

**Bài 5.3:** Các phép toán với set:
- Hợp (union): A ∪ B
- Giao (intersection): A ∩ B
- Hiệu (difference): A - B
- Hiệu đối xứng (symmetric difference): A △ B

````python
A = {1,2,3,4,5}
B = {4,5,6,7,8}
#Hợp
hop = A.union(B)
print(f"Hợp: {hop}")

#Giao
giao = A.intersection(B)
print(f"Giao: {giao}")

#Hiệu
hieu = A.difference(B)
print(f"Hiệu: {hieu}")

#Hiệu đối xứng
hieudx = A.symmetric_difference(B)
print(f"Hiệu đối xứng: {hieudx}")
````

**Bài 5.4:** Kiểm tra subset và superset

````python
setcha = {1,2,3,4,5}
setcon = {1,2,3}

subset = setcon.issubset(setcha)
print(f"Set con có phải là tập con của set cha hay không? {subset}")

superset = setcha.issuperset(setcon)
print(f"Set cha có phải là tập cha của set con hay không? {superset}")
````

### CẤP ĐỘ 3: NÂNG CAO

**Bài 6.8:** Group các sinh viên theo điểm số

Input: `[{'name': 'An', 'score': 8}, {'name': 'Binh', 'score': 7}, {'name': 'Chi', 'score': 8}]`

Output: `{8: ['An', 'Chi'], 7: ['Binh']}`

````python
inputs = [
    {'name': 'An', 'score': 8}, 
    {'name': 'Bình', 'score': 7}, 
    {'name': 'Chi', 'score': 8}
]

output = {}

for sv in inputs:
    diem = sv['score']
    ten = sv['name']

    if diem not in output:
        output[diem] = []
    
    output[diem].append(ten)
print(output)
````

**Bài 6.9:** Nested Dictionary - Tạo từ điển chứa thông tin học sinh với nhiều môn học

Ví dụ:
```python
{
    'An': {'Toán': 8, 'Lý': 7, 'Hóa': 9},
    'Bình': {'Toán': 9, 'Lý': 8, 'Hóa': 7}
}
```

Tính điểm trung bình của mỗi học sinh

````python
students = {
    'An': {'Toán': 8, 'Lý': 7, 'Hóa': 9},
    'Bình': {'Toán': 9, 'Lý': 8, 'Hóa': 7}
}
for name, subjects in students.items():
    scores = subjects.values()
    average = sum(scores) / len(scores)
    print(f"Điểm trung bình của {name}: {average:.2f}")
````

**Bài 7.1: Quản lý thư viện sách**

Tạo chương trình quản lý thư viện với các chức năng:
- Thêm sách (tên, tác giả, năm xuất bản)
- Xóa sách
- Tìm sách theo tên
- Liệt kê tất cả sách
- Tìm sách theo tác giả

````python
thu_vien = []

def them_sach(ten, tac_gia, nam):
    sach = {"ten": ten, "tac_gia": tac_gia, "nam": nam}
    thu_vien.append(sach)
    print(f"Đã thêm sách: {ten}")

def xoa_sach(ten):
    global thu_vien
    truoc_khi_xoa = len(thu_vien)
    thu_vien = [s for s in thu_vien if s["ten"].lower() != ten.lower()]
    if len(thu_vien) < truoc_khi_xoa:
        print(f"Đã xóa sách: {ten}")
    else:
        print("Không tìm thấy sách để xóa")

def tim_theo_ten(ten):
    ket_qua = [s for s in thu_vien if ten.lower() in s["ten"].lower()]
    return ket_qua

def tim_theo_tac_gia(tac_gia):
    ket_qua = [s for s in thu_vien if tac_gia.lower() in s["tac_gia"].lower()]
    return ket_qua

def liet_ke_sach():
    if not thu_vien:
        print("Thư viện trống.")
    for s in thu_vien:
        print(f"Tên: {s['ten']} | Tác giả: {s['tac_gia']} | Năm: {s['nam']}")

them_sach("Lập trình Python", "Nguyễn Văn A", 2023)
them_sach("Dữ liệu cấu trúc", "Trần Thị B", 2022)
liet_ke_sach()
````

**Bài 7.2: Phân tích dữ liệu bán hàng**

Cho dữ liệu bán hàng:
```python
sales = [
    {'product': 'Laptop', 'quantity': 2, 'price': 20000000},
    {'product': 'Phone', 'quantity': 5, 'price': 10000000},
    {'product': 'Laptop', 'quantity': 1, 'price': 20000000},
]
```

Tính:
- Tổng doanh thu
- Sản phẩm bán chạy nhất
- Tổng doanh thu theo từng sản phẩm

````python
sales = [
    {'product': 'Laptop', 'quantity': 2, 'price': 20000000},
    {'product': 'Phone', 'quantity': 5, 'price': 10000000},
    {'product': 'Laptop', 'quantity': 1, 'price': 20000000},
]

#tính tổng doanh thu
tong_doanh_thu = sum(item['quantity'] * item['price'] for item in sales)

#tính tổng doanh thu theo từng sản phẩm
stats = {}
for item in sales:
    p = item['product']
    rev = item['quantity'] * item['price']
    qty = item['quantity']
    
    if p not in stats:
        stats[p] = {'revenue': 0, 'total_qty': 0}
    stats[p]['revenue'] += rev
    stats[p]['total_qty'] += qty

#tìm sản phẩm bán chạy nhất (dựa trên số lượng)
san_pham_best_seller = max(stats, key=lambda x: stats[x]['total_qty'])

print(f"Tổng doanh thu: {tong_doanh_thu:,} VNĐ")
print(f"Sản phẩm bán chạy nhất: {san_pham_best_seller}")
print("Tổng doanh thu theo từng sản phẩm:")
for p, data in stats.items():
    print(f" - {p}: {data['revenue']:,} VNĐ")
````
