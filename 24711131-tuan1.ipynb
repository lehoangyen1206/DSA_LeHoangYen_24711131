```python
def KiemTraSoNguyenTo(n):
    if n<1:
        return False
    for i in range(2,int(n*0.5) +1):
        if n%i == 0:
            return False
    return True
n=4
kq = KiemTraSoNguyenTo(n)
if kq == False:
    print(f"{n} không là SNT")
else:
    print(f"{n} là SNT")
```
