Sinh viên bao gồm mã sinh viên, họ tên, địa chỉ, điểm csdl, điểm big data

```python
class SinhVien:
    def __init__(self, maSV, HoTen, DiaChi, DiemCSDL, DiemBigData):
        self.maSV = maSV
        self.HoTen = HoTen
        self.DiaChi = DiaChi
        self.DiemCSDL = DiemCSDL
        self.DiemBigData = DiemBigData
    def TinhTrungBinh(self):
        return (self.DiemBigData + self.DiemCSDL)/2
```

```python
sv1 = SinhVien('sv01', 'Yến', 'Bình Chánh', 8, 9)
sv1.TinhTrungBinh()
```

```python
class SinhVienKSTN (SinhVien):
    def __init__(self, maSV, HoTen, DiaChi, DiemCSDL, DiemBigData, DiemEng):
        super().__init__(self, maSV, HoTen, DiaChi, DiemCSDL, DiemBigData)
        self.DiemEng = DiemEng

    def TinhTrungBinh(self):
        return (self.DiemBigData + self.DiemCSDL +self.DiemEng)/3
        #return super().TinhTrungBinh + self.DiemEng
```
