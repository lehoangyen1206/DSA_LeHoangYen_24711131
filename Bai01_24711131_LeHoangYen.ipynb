## **III. Data init**

````python
# Danh sách sinh viên ban đầu
students = [
    {
        "id": "SV001",
        "name": "Nguyễn Văn An",
        "age": 20,
        "scores": {"SAD": 8.5, "DSA": 7.0, "BigData": 9.0}
    },
    {
        "id": "SV002",
        "name": "Trần Thị Bình",
        "age": 21,
        "scores": {"SAD": 6.5, "DSA": 7.5, "BigData": 6.0}
    },
    {
        "id": "SV003",
        "name": "Lê Văn Cường",
        "age": 19,
        "scores": {"SAD": 9.0, "DSA": 8.5, "BigData": 9.5}
    },
    {
        "id": "SV004",
        "name": "Phạm Thị Dung",
        "age": 20,
        "scores": {"SAD": 5.0, "DSA": 4.5, "BigData": 6.0}
    },
    {
        "id": "SV005",
        "name": "Hoàng Văn Em",
        "age": 22,
        "scores": {"SAD": 7.0, "DSA": 7.0, "BigData": 7.5}
    }
]


print(f"Số lượng sinh viên: {len(students)}")
````

````python
def calculate_average(scores):
    """
    Tính điểm trung bình từ dictionary điểm
    
    Args:
        scores (dict): Dictionary chứa điểm các môn
    
    Returns:
        float: Điểm trung bình
    """
    if not scores:
        return 0.0
    total = sum(scores.values())
    return total / len(scores)
print(f"Điểm trung bình: {calculate_average(students[2]['scores']):.2f}")
````

## **IV. CÁC CHỨC NĂNG**

### 1: Hiển thị danh sách sinh viên

````python
def display_all_students():
    """
    Hiển thị danh sách tất cả sinh viên
    In ra: ID, Tên, Tuổi, Điểm trung bình
    """
    print(f"{'ID':<6} {'Tên':<20} {'Tuổi':<6} {'Điểm TB'}")
    print("-" * 45)

    for student in students:
        print(
            f"{student['id']:<6} "
            f"{student['name']:<20} "
            f"{student['age']:<6} "
            f"{avg:.2f}"
        )
display_all_students
````

### 2: Thêm sinh viên mới

````python
def add_new_student():
    """
    Thêm sinh viên mới
    - Tự động tạo ID theo format SV001, SV002...
    - Nhập: name, age
    - Nhập điểm 3 môn: SAD, DSA, BigData
    """
    print("\n--- THÊM SINH VIÊN MỚI ---")
    
    new_id = f"SV{len(students) + 1:03d}"
    name = input("Nhập tên sinh viên: ")
    age = int(input("Nhập tuổi: "))
    scores = {
        "SAD": float(input("Nhập điểm SAD: ")),
        "DSA": float(input("Nhập điểm DSA: ")),
        "BigData": float(input("Nhập điểm BigData: "))
    }
    
    student = {
        "id": new_id,
        "name": name,
        "age": age,
        "scores": scores
    }
    students.append(student)
    
add_new_student()
````

### 3: Cập nhật điểm

````python
def update_scores():
    """
    Cập nhật điểm sinh viên
    - Nhập ID sinh viên
    - Hiển thị điểm hiện tại
    - Cho phép cập nhật từng môn (Enter để bỏ qua)
    - Cập nhật dictionary scores
    """
    print("\n--- CẬP NHẬT ĐIỂM ---")

    student_id = input("Nhập ID sinh viên: ").strip()

    for student in students:
        if student["id"] == student_id:
            print(f"\nSinh viên: {student['name']}")
            print("Điểm hiện tại:")

            for subject, score in student["scores"].items():
                print(f"  {subject}: {score}")

            for subject in student["scores"]:
                new_score = input(
                    f"Nhập điểm mới cho {subject} (Enter để bỏ qua): "
                ).strip()

                if new_score != "":
                    student["scores"][subject] = float(new_score)

            print("Cập nhật điểm thành công!")
            return

    print("Không tìm thấy sinh viên có ID:", student_id)
update_scores()

````

### 4: Xóa sinh viên theo ID

````python
def delete_student():
    """
    Xóa sinh viên theo ID 
    """
    print("\n--- XÓA SINH VIÊN ---")
    find_id = input("Nhập ID sinh viên cần xóa: ").strip()
    
    for student in students:
        if student["id"] == find_id:
            students.remove(student)
            print("Đã xóa sinh viên có ID:", find_id)
            return

    print("Không tìm thấy sinh viên có ID:", find_id)
delete_student()
````

### 5: Tìm sinh viên theo tên

````python
def search_by_name():
    """
    Tìm sinh viên theo tên (không phân biệt hoa thường, tìm kiếm gần đúng)
    """
    print("\n--- TÌM SINH VIÊN THEO TÊN ---")
    keyword = input("Nhập tên cần tìm: ").strip().lower()

    if not keyword:
        print("Vui lòng không để trống tên cần tìm.")
        return

    found = False
    header = f"{'ID':<6} {'Họ và Tên':<20} {'Tuổi':<6} {'ĐTB':<5}"
    print("-" * len(header))
    print(header)
    print("-" * len(header))

    for student in students:
        if keyword in student["name"].lower():
            avg = calculate_average(student["scores"])
            
            print(
                f"{student['id']:<6} "
                f"{student['name']:<20} "
                f"{student['age']:<6} "
                f"{avg:<5.2f}"
            )
            found = True

    if not found:
        print("Không tìm thấy sinh viên phù hợp.")
    else:
        print("-" * len(header))
def calculate_average(scores):
    if not scores: return 0
    return sum(scores) / len(scores)
search_by_name()
````

### 6: Xếp hạng sinh viên

````python
def rank_students():
    """
    Xếp hạng sinh viên theo điểm trung bình
    - Tính điểm TB cho tất cả sinh viên
    - Sắp xếp theo điểm TB giảm dần
    - Sử dụng sorted() với key=lambda
    - Hiển thị bảng xếp hạng có số thứ tự
    """
    print("\n--- BẢNG XẾP HẠNG SINH VIÊN ---")
    
    def rank_students():
        print("\n--- BẢNG XẾP HẠNG SINH VIÊN ---")

    ranked_students = sorted(
        students,
        key=lambda s: calculate_average(s["scores"]),
        reverse=True
    )

    print(f"{'Hạng':<5} {'ID':<6} {'Tên':<20} {'Điểm TB'}")
    print("-" * 45)

    for i, student in enumerate(ranked_students, start=1):
        avg = calculate_average(student["scores"])
        print(
            f"{i:<5} "
            f"{str(student['id']):<6} "
            f"{student['name']:<20} "
            f"{avg:.2f}"
        )
rank_students()
````

### 7: Thống kê theo mức điểm

````python
def statistics_by_grade():
    """
    Thống kê sinh viên theo mức điểm
    - Giỏi: điểm TB >= 8.0
    - Khá: 6.5 <= điểm TB < 8.0
    - Trung bình: 5.0 <= điểm TB < 6.5
    - Yếu: điểm TB < 5.0
    - Đếm số sinh viên mỗi loại
    """
    print("\n--- THỐNG KÊ THEO MỨC ĐIỂM ---")
    
    gioi = kha = trungbinh = yeu = 0
    
    for student in students:
        avg = calculate_average(student["scores"])
        if avg >= 8.0:
            gioi += 1
        elif avg >= 6.5:
            kha += 1
        elif avg >= 5.0:
            trungbinh += 1
        else:
            yeu += 1
    print(f"Số sinh viên Giỏi: {gioi}")
    print(f"Số sinh viên Khá: {kha}")
    print(f"Số sinh viên Trung bình: {trungbinh}")
    print(f"Số sinh viên Yếu: {yeu}")
        
statistics_by_grade()
````

### 8: Tìm sinh viên điểm cao nhất/thấp nhất

````python
def find_top_bottom_students():
    """
    Tìm sinh viên có điểm cao nhất và thấp nhất
    - Sử dụng max()/min() với key
    - Hiển thị thông tin chi tiết
    """
    print("\n--- SINH VIÊN ĐIỂM CAO NHẤT VÀ THẤP NHẤT ---")

    if not students:
        print("Danh sách sinh viên trống")
        return

    # Sinh viên điểm cao nhất
    top_student = max(
        students,
        key=lambda s: calculate_average(s["scores"])
    )

    # Sinh viên điểm thấp nhất
    bottom_student = min(
        students,
        key=lambda s: calculate_average(s["scores"])
    )

    top_avg = calculate_average(top_student["scores"])
    bottom_avg = calculate_average(bottom_student["scores"])

    print("\nSinh viên điểm CAO nhất:")
    print(f"ID: {top_student['id']}")
    print(f"Tên: {top_student['name']}")
    print(f"Tuổi: {top_student['age']}")
    print(f"Điểm TB: {top_avg:.2f}")

    print("\nSinh viên điểm THẤP nhất:")
    print(f"ID: {bottom_student['id']}")
    print(f"Tên: {bottom_student['name']}")
    print(f"Tuổi: {bottom_student['age']}")
    print(f"Điểm TB: {bottom_avg:.2f}")
   
find_top_bottom_students()
````

````python
def main_menu():
    while True:
        print("\n" + "="*50)
        print("     CHƯƠNG TRÌNH QUẢN LÝ SINH VIÊN VÀ ĐIỂM")
        print("="*50)
        print("1. Hiển thị danh sách sinh viên")
        print("2. Thêm sinh viên mới")
        print("3. Cập nhật điểm")
        print("4. Xóa sinh viên theo ID")
        print("5. Tìm sinh viên theo tên")
        print("6. Xếp hạng sinh viên")
        print("7. Thống kê theo mức điểm")
        print("8. Tìm sinh viên điểm cao nhất/thấp nhất")
        print("0. Thoát")
        print("="*50)
        
        choice = input("Chọn chức năng (0-8): ").strip()

        if choice == "1":
            display_all_students()

        elif choice == "2":
            add_new_student()

        elif choice == "3":
            update_scores()

        elif choice == "4":
            delete_student()

        elif choice == "5":
            search_by_name()

        elif choice == "6":
            rank_students()

        elif choice == "7":
            statistics_by_grade()

        elif choice == "8":
            find_top_bottom_students()

        elif choice == "0":
            print("Thoát chương trình")
            break

        else:
            print("\nLựa chọn không hợp lệ! Vui lòng chọn lại")

main_menu()
````
