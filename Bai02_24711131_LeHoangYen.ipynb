## II. CẤU TRÚC DỮ LIỆU

Mỗi cuốn sách được lưu trữ dưới dạng Dictionary với cấu trúc:

```python
{
    "id": 1,
    "title": "Tên sách",
    "author": "Tác giả",
    "year": 2024,
    "status": "available"  # available hoặc borrowed
}
```

````python
books = [
    {"id": 1, "title": "Lập trình Python cơ bản", "author": "Nguyễn Văn A", "year": 2020, "status": "available"},
    {"id": 2, "title": "Khoa học dữ liệu với Python", "author": "Trần Thị B", "year": 2021, "status": "borrowed"},
    {"id": 3, "title": "Machine Learning cơ bản", "author": "Lê Văn C", "year": 2022, "status": "available"},
    {"id": 4, "title": "Python nâng cao", "author": "Nguyễn Văn A", "year": 2023, "status": "available"},
    {"id": 5, "title": "Big Data với Python", "author": "Phạm Thị D", "year": 2024, "status": "borrowed"}
]

print("Đã khởi tạo dữ liệu mẫu thành công!")
print(f"Số lượng sách: {len(books)}")
````

````python
def display_all_books():
    """
    Hiển thị danh sách tất cả các sách trong thư viện
    In ra: ID, Tên sách, Tác giả, Năm xuất bản, Trạng thái
    """
    print(f"{'ID':<6} {'Tên sách':<30} {'Tác giả':<20} {'Năm xuất bản':<15} {'Trạng thái':<20}")
    print("-"*85)

    for book in books:
        print(
            f"{book['id']:<6} "
            f"{book['title']:<30} "
            f"{book['author']:<20} "
            f"{book['year']:<15} "
            f"{book['status']:<20} "
        )
display_all_books()
````

````python
def xoa_sach():
    find_id = input("Nhập ID sách cần xóa:")

    for book in books:
        if book["id"] == find_id:
            books.remove(book)
            print("Đã xóa sách có ID:", find_id)
            return

    print("Không tìm thấy sách có ID:", find_id)
xoa_sach()
````

````python
def add_new_book():
    """
    Thêm sách mới vào thư viện
    - Tự động tạo ID
    - lấy giá trị truyền vô: title, author, year
    - Mặc định status = "available"
    """
    print("\n--- THÊM SÁCH MỚI ---")
    new_id = f"{len(books) +1:01d}"
    title = input("Nhập tên sách mới: ")
    author = input("Nhập tên tác giả: ")
    year = int(input("Nhập năm xuất bản: "))
    status = input("Nhập trạng thái: ")

    book ={
        "id": new_id,
        "title": title,
        "author": author,
        "year": year,
        "status": status
    }
    books.append(book)
add_new_book()
````

````python
def borrow_book():
    """
    Mượn sách từ thư viện
    """
    print("\n--- MƯỢN SÁCH ---")
    find_id = input("Nhập ID sách cần mượn: ")
    
    found = False

    for book in books:
        if book["id"] == find_id.strip():
            found = True
            if book["status"] == "available":
                book["status"] = "borrowed"
                print(f"Bạn đã mượn thành công cuốn: {book['title']}")
            else:
                print(f"Cuốn sách '{book['title']}' hiện đã có người mượn")
            break
    if not found:
        print("Không tìm thấy sách với ID này trong hệ thống")
borrow_book()
    
````

````python
def return_book():
    """
    Trả sách cho thư viện
    - Nhập ID sách cần trả
    - Kiểm tra status = "borrowed"
    - Cập nhật status = "available"
    - Thông báo kết quả
    """
    print("\n--- TRẢ SÁCH ---")
    find_id = input("Nhập ID sách cần trả: ").strip()
    
    found = False

    for book in books:
        if str(book["id"]) == find_id:
            found = True
            
            if book["status"] == "borrowed":
                book["status"] = "available"
                print(f"Thành công: Bạn đã trả cuốn '{book['title']}'")
            else:
                print(f"Cuốn '{book['title']}' hiện đang có sẵn trong kho, không cần trả")

    if not found:
        print(f"Không tìm thấy sách có ID '{find_id}' trong hệ thống")

return_book()
    
  
````

````python
def search_by_author():
    """
    Tìm kiếm sách theo tác giả
    - Nhập tên tác giả
    - Duyệt list và hiển thị tất cả sách của tác giả đó
    - Sử dụng lower() để không phân biệt hoa thường
    """
    print("\n--- TÌM SÁCH THEO TÁC GIẢ ---")
    
    keyword = input("Nhập tên tác giả cần tìm: ").strip().lower()

    found = False

    for book in books:
        if keyword in book["author"].lower():
            print(
                f"{book['id']:<6} "
                f"{book['title']:<20} "
                f"{book['author']:<20} " 
                f"{book['year']:<6} "
                f"{book['status']:<20} "
            )
            found = True
    if not found:
         print("Không tìm thấy sinh viên phù hợp")

search_by_author()
````

````python
def filter_by_year():
    """
    Lọc sách theo khoảng năm xuất bản
    - Nhập năm bắt đầu và năm kết thúc
    - Hiển thị các sách trong khoảng năm đó
    """
    print("\n--- LỌC SÁCH THEO NĂM XUẤT BẢN ---")
    
    try:
        start_year = int(input("Nhập năm bắt đầu: "))
        end_year = int(input("Nhập năm kết thúc: "))
        
        print(f"\nDanh sách sách từ năm {start_year} đến {end_year}:")
        print("-" * 50)
        print(f"{'ID':<5} {'Tên sách':<25} {'Năm XB':<10}")
        print("-" * 50)
        
        found = False
        for book in books:
            if start_year <= book["year"] <= end_year:
                print(f"{book['id']:<5} {book['title']:<25} {book['year']:<10}")
                found = True
        
        if not found:
            print("Không tìm thấy sách nào trong khoảng năm này.")
            
    except ValueError:
        print("Vui lòng chỉ nhập số nguyên cho năm xuất bản!")

filter_by_year()
    
````

````python
def show_statistics():
    """
    Hiển thị thống kê thư viện
    - Tổng số sách
    - Số sách đang available
    - Số sách đang borrowed
    - Tính phần trăm
    """
    print("\n--- THỐNG KÊ THƯ VIỆN ---")
    
    total_books = len(books)
    if total_books == 0:
        print("Thư viện hiện chưa có sách nào.")
        return

    available_count = sum(1 for book in books if book["status"] == "available")
    borrowed_count = sum(1 for book in books if book["status"] == "borrowed")

    available_pct = (available_count / total_books) * 100
    borrowed_pct = (borrowed_count / total_books) * 100

    print(f"Tổng số sách trong kho: {total_books}")
    print("-" * 30)
    print(f"Sách sẵn có (Available): {available_count} cuốn ({available_pct:.1f}%)")
    print(f"Sách đã mượn (Borrowed) : {borrowed_count} cuốn ({borrowed_pct:.1f}%)")
    print("-" * 30)
show_statistics()
````

````python
def main_menu():
    while True:
        print("\n" + "="*25)
        print("   QUẢN LÝ THƯ VIỆN")
        print("="*25)
        print("1. Hiện tất cả sách")
        print("2. Thêm sách mới")
        print("3. Mượn sách")
        print("4. Trả sách")
        print("5. Tìm sách theo tác giả")
        print("6. Lọc sách theo năm")
        print("7. Xem thống kê (Tổng/Sẵn/Mượn)")
        print("0. Thoát chương trình")
        
        choice = input("\nChọn chức năng (0-7): ")
        
        if choice == "1": show_all_books()
        elif choice == "2": add_book()
        elif choice == "3": borrow_book()
        elif choice == "4": return_book()
        elif choice == "5": search_by_author()
        elif choice == "6": filter_by_year()
        elif choice == "7": show_statistics()
        elif choice == "0":
            print("Tạm biệt!")
            break
        else:
            print("Lựa chọn không hợp lệ, vui lòng nhập lại.")
````

````python
main_menu()
````
